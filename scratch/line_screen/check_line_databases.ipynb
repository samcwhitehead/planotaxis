```python
import cPickle
import os
```

```python
# load in the line database -- code taken from 'muscle_model.py'
database_path = '/home/sam/Documents/code/muscle_imaging/muscle_model/line_database.cpkl'
f = open(database_path,'rb')
line_database = cPickle.load(f)
f.close()
```

```python
# print results in a way that's interpretable
for line in line_database.keys():
    print('\n%s muscle data:'%(line))
    line_dict = line_database[line]
    
    for muscle in line_dict.keys():
        print('      %s : %1.1f' %(muscle, line_dict[muscle]))
```
